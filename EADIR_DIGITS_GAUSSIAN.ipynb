#K-means classfier
The k-means algorithm is a popular unsupervised machine learning algorithm used for clustering data. The goal of the algorithm is to partition a set of data points into k clusters, where each point belongs to the cluster with the nearest mean, or centroid.

The k-means algorithm works by first randomly selecting k points from the dataset to act as the initial centroids for the clusters. Then, for each data point in the dataset, the algorithm calculates the distance to each centroid and assigns the point to the cluster whose centroid is closest. After all data points have been assigned to a cluster, the algorithm recomputes the centroid of each cluster based on the mean of the points in the cluster. This process of assigning points to clusters and recomputing centroids is repeated iteratively until the clusters no longer change.

The k-means algorithm has many applications, including image segmentation, anomaly detection, and customer segmentation. However, it is important to note that the algorithm is sensitive to the initial choice of centroids, and may converge to a suboptimal solution. As such, it is common practice to run the algorithm multiple times with different initial centroids and choose the clustering with the lowest objective function value.

The k-means algorithm can efficiently assign labels to clusters that closely resemble human intuition for simple data blobs. For instance, visually inspecting the data would likely result in a similar labeling.

````python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
````

````python
# Generate some data
from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=400, centers=4,
                       cluster_std=0.60, random_state=0)
````

````python
# Plot the data with K Means Labels
from sklearn.cluster import KMeans

kmeans = KMeans(4, random_state=0)
labels = kmeans.fit(X).predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis');
````

The k-means model can be conceptualized as a circle (or hyper-sphere in higher dimensions) at the centroid of each cluster, with a radius determined by the farthest point in the cluster. This radius serves as a strict boundary for assigning data points to the corresponding cluster during training: any point beyond this boundary is excluded from the cluster. The following function can be used to visualize this clustering model:

````python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

def plot_kmeans(kmeans, X, n_clusters=4, rseed=0, ax=None):
    labels = kmeans.fit_predict(X)

    # plot the input data
    ax = ax or plt.gca()
    ax.axis('equal')
    ax.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis', zorder=2)

    # plot the representation of the KMeans model
    centers = kmeans.cluster_centers_
    radii = [cdist(X[labels == i], [center]).max()
             for i, center in enumerate(centers)]
    for c, r in zip(centers, radii):
        ax.add_patch(plt.Circle(c, r, fc='#CCCCCC', lw=3, alpha=0.5, zorder=1))
````

````python
kmeans = KMeans(n_clusters=4, random_state=0)
plot_kmeans(kmeans, X)
````

In the above function, we have the kmeans classifier `kmeans = KMeans(n_clusters=4, random_state=0)` and it is called `labels = kmeans.fit_predict(X)` to generate labels.



One crucial limitation of the k-means algorithm is that it assumes circular cluster shapes and cannot handle oblong or elliptical clusters without additional modifications. Therefore, if we were to apply a transformation to the same dataset, the resulting cluster assignments could become ambiguous and less accurate.

# Generalizing E–M: Gaussian Mixture Models

Gaussian Mixture Models (GMMs) are a probabilistic unsupervised machine learning algorithm used for clustering and density estimation of data. A GMM represents a dataset as a mixture of several Gaussian distributions, where each Gaussian represents a cluster in the data. Unlike k-means, GMMs do not assume that the clusters have a circular shape or are of equal size, and can account for oblong or elliptical clusters.

The GMM algorithm works by first randomly initializing the parameters of the Gaussian distributions, including their means, variances, and mixing coefficients. Then, for each data point in the dataset, the algorithm computes the probability of the point belonging to each Gaussian distribution. These probabilities are used to update the parameters of the Gaussian distributions using the Expectation-Maximization (EM) algorithm. The EM algorithm iteratively maximizes the likelihood of the data given the Gaussian mixture model, until convergence is achieved.

````python
from sklearn.mixture import GaussianMixture as GMM
gmm = GMM(n_components=4).fit(X)
labels = gmm.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis');
````

But because GMM contains a probabilistic model under the hood, it is also possible to find probabilistic cluster assignments—in Scikit-Learn this is done using the predict_proba method. This returns a matrix of size [n_samples, n_clusters] which measures the probability that any point belongs to the given cluster:

````python
probs = gmm.predict_proba(X)
print(probs[:5].round(3))
````

````python
from matplotlib.patches import Ellipse

def draw_ellipse(position, covariance, ax=None, **kwargs):
    """Draw an ellipse with a given position and covariance"""
    ax = ax or plt.gca()
    
    # Convert covariance to principal axes
    if covariance.shape == (2, 2):
        U, s, Vt = np.linalg.svd(covariance)
        angle = np.degrees(np.arctan2(U[1, 0], U[0, 0]))
        width, height = 2 * np.sqrt(s)
    else:
        angle = 0
        width, height = 2 * np.sqrt(covariance)
    
    # Draw the Ellipse
    for nsig in range(1, 4):
        ax.add_patch(Ellipse(position, nsig * width, nsig * height,
                             angle, **kwargs))
        
def plot_gmm(gmm, X, label=True, ax=None):
    ax = ax or plt.gca()
    labels = gmm.fit(X).predict(X)
    if label:
        ax.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis', zorder=2)
    else:
        ax.scatter(X[:, 0], X[:, 1], s=40, zorder=2)
    ax.axis('equal')
    
    w_factor = 0.2 / gmm.weights_.max()
    for pos, covar, w in zip(gmm.means_, gmm.covariances_, gmm.weights_):
        draw_ellipse(pos, covar, alpha=w * w_factor)
````

````python
gmm = GMM(n_components=4, random_state=42)
plot_gmm(gmm, X)
````

Note that gmm is trained with `gmm = gmm.fit(X)` in the `plot_gmm` function.

# PCA

PCA, or Principal Component Analysis, is a technique used for dimensionality reduction and data compression. It is a popular method for reducing the number of features in a dataset while preserving the majority of the information contained in the original data.

PCA works by finding a new set of orthogonal axes, called principal components, that represent the largest sources of variation in the data. The first principal component is the direction that explains the most variance in the data, and each subsequent principal component explains as much of the remaining variance as possible. These principal components are sorted in order of importance, with the first principal component being the most important.



````python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import norm, svd, eig
from numpy.random import rand, randn
N = 5; L = 100
X = randn(N,L); plt.plot(X[0,:],X[1,:],'.r')
````

````python
X.shape
````

Given random data X, we first compute its Expectation:

````python
E_x = np.average(X,axis=1); print('E[x]:',E_x)
````

The centered data can be expressed as:

````python
Exx = E_x.reshape(5,1); ones = np.ones(L).reshape(1,L)
Exrep = np.dot(Exx,ones)
X_c = X-Exrep
X_c
````

The correlation matrix $C_x\inm{N}{N}$ for a random variable $\rv{x}\inv{N}$ presents all scalar correlations obtained for any pair $(i,j)$ of  its components $\rv{x}[i], \rv{x}[j]\in\bb{R}$:
$$
(C_x)_{ij} \eqd \bb{E}\left[\rv{x}[i]\cdot \rv{x}[j]\right],\quad i,j=1,\dots,n
$$
Correlation matrix can be found:

````python
L = X.shape[1]
C_x = np.dot(X,X.T)/L
print('correlation matrix:\n',C_x)
````

The covariance matrix $R_x$ for a random variable $r_x\inv{N}$ by the definition equals to the correlation matrix $C_y$ for $\rv{y}\eqd\rv{x} - \bb{E}[\rv{x}]$ $-$ the centered version of $\rv{x}.$

````python
R_x = np.dot(X_c,X_c.T)/L
print('covariance matrix:\n',R_x)
````

````python
eigenvalues, eigenvectors = eig(R_x);
````

# Precision, Recall, F1 score and Confusion Matrix


Recall, precision, and F1 score are common performance metrics used in classification problems. These metrics provide a way to evaluate how well a classifier is performing and how accurate it is in predicting the class labels of a dataset.

Recall measures the proportion of actual positive samples that are correctly identified as positive by the classifier. Mathematically, it is defined as the ratio of true positives to the sum of true positives and false negatives. A high recall value indicates that the classifier is correctly identifying most of the positive samples in the dataset.

Precision, on the other hand, measures the proportion of positive predictions that are actually correct. It is defined as the ratio of true positives to the sum of true positives and false positives. A high precision value indicates that the classifier is making fewer false positive predictions and is accurately predicting the positive samples.

The F1 score is a weighted average of recall and precision, providing a single score that summarizes the overall performance of the classifier. It is defined as the harmonic mean of recall and precision and takes both metrics into account. The F1 score is a useful metric when there is an uneven distribution of positive and negative samples in the dataset, as it provides a balance between recall and precision.

````python
from sklearn.metrics import precision_score, recall_score, f1_score , accuracy_score

# Example true labels and predicted labels
y_true = [0, 1, 1, 0, 1, 0, 0, 1, 0, 1]
y_pred = [1, 1, 0, 0, 1, 0, 1, 1, 0, 0]

# Calculate precision, recall, and F1 score
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

# Print the results
print("Precision: {:.2f}".format(precision))
print("Recall: {:.2f}".format(recall))
print("F1 Score: {:.2f}".format(f1))
````

A confusion matrix is a table used to evaluate the performance of a classification algorithm. It compares the predicted labels of a model to the true labels of a dataset and shows how many samples were classified correctly or incorrectly for each class.

The confusion matrix is typically a square matrix, where the rows represent the true labels and the columns represent the predicted labels. Each entry in the matrix shows the count of the corresponding pair of true and predicted labels. The diagonal elements represent the correctly classified samples, while the off-diagonal elements represent the misclassified samples.

From the confusion matrix, several performance metrics can be calculated, including accuracy, precision, recall, and F1 score. These metrics are useful for evaluating the effectiveness of a classification algorithm, identifying which classes the model struggles with, and fine-tuning the model parameters to improve performance.

Overall, a confusion matrix is an essential tool for evaluating the performance of a classification algorithm and gaining insights into its strengths and weaknesses.

````python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import numpy as np

# Example true labels and predicted labels
y_true = np.array([0, 1, 0, 1, 1, 0, 1, 0, 1, 0])
y_pred = np.array([0, 0, 1, 1, 1, 0, 1, 1, 1, 0])

# Compute the confusion matrix
cm = confusion_matrix(y_true, y_pred)
````

````python
def plot_confusion_matrix(cm):
  # Plot the confusion matrix
  fig, ax = plt.subplots()
  im = ax.imshow(cm, cmap='Blues')

  # Add colorbar
  cbar = ax.figure.colorbar(im, ax=ax)

  # Set axis labels
  ax.set_xlabel('Predicted Label')
  ax.set_ylabel('True Label')

  # Add title
  ax.set_title('Confusion Matrix')

  # Add tick marks
  tick_marks = np.arange(len(np.unique(y_true)))
  ax.set_xticks(tick_marks)
  ax.set_yticks(tick_marks)
  ax.set_xticklabels(np.unique(y_true))
  ax.set_yticklabels(np.unique(y_true))

  # Add text annotations
  thresh = cm.max() / 2.
  for i in range(cm.shape[0]):
      for j in range(cm.shape[1]):
          ax.text(j, i, format(cm[i, j], 'd'),
                  ha="center", va="center",
                  color="white" if cm[i, j] > thresh else "black")

  plt.show()

plot_confusion_matrix(cm)
````
