###*Kontekst importowania narzędzi modyfikowanych w trakcie sesji Colaba*<br>Context of importing tools being  modified within the Colab session

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/e/e9/Flag_of_Poland_%28normative%29.svg/240px-Flag_of_Poland_%28normative%29.svg.png"  width="20px"/><br>
W Pythonie, dla oszczędności czasu ponowne wykonanie polecenia `import any_object` dotyczy tylko reprezentacji tego modułu w pamięci RAM, interpreter nie sięga już do zasobów dyskowych. Czasem, np. gdy dokonaliśmy zmian obiektu świadomie lub za pośrednictwem używanej aplikacji, to konieczne jest wymuszenie ponownego jego importu. By uniknąć restartu całej aplikacji używamy funkcji `reload` z modułu `importlib`:  

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/83/Flag_of_the_United_Kingdom_%283-5%29.svg/240px-Flag_of_the_United_Kingdom_%283-5%29.svg.png" width="20px"/><br>
For the sake of time savings,  the second repeated execution of command `import any_object` refers to the already imported module already available in RAM, so the interpreter is not checking on update of the module on disc. In order to force it to do the import again without restarting of the whole application the command `reload` from the module `importlib` is used:  


````
from importlib import reload  # optional
  # PL: przyda się, gdy w trakcie sesji Colaba będziesz modyfikować swoje pliki Pythona
  # EN: useful if your Python files are modified during the Colab session
````

###Informacja o używanym archiwum obrazów<br>Information on image archive to be used 

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/e/e9/Flag_of_Poland_%28normative%29.svg/240px-Flag_of_Poland_%28normative%29.svg.png"  width="20px"/>

W typowym projekcie DNN archiwum przykładów stosowanych w budowie i testowaniu modelu nie wymaga konwersji na inny typ kompresji obrazów, czy też inny typ samego archiwum.

W tym przypadku, archiwum Kaggle/MNIST ma postać dwóch arkuszy kalkulacyjnych CSV $-$ jeden z przykładami uczącymi, a drugi z przykładami testowymi. Arkusz CSV, po wierszu nagłówkowym, ma w kolejnych wierszach:
- w pierwszej kolumnie etykietę obrazu, tj. numeryczny identyfikator klasy obrazu $-$ tutaj jest to symbol cyfry, który ten obraz reprezentuje,
- w kolejnych $28\cdot 28=784$ kolumnach wartości wszystkich pikseli obrazu

Chcemy utworzyć archiwum ZIP na bazie folderu, w którym znajdą się wszystkie obrazy z archiwum CSV w formatach PNG, JPG oraz RAW, tj. w reprezentacji zawierającej dokładnie $784$ bajty.

Ponieważ w sieci znajdują się różne archiwa używające nazwy MNIST dla przykładów zaczerpnietych ze zbioru danych utworzonego przez amerkańskie biuro standaryzacji NIST, dla ustalenia, o które archiwum MNIST nam chodzi, zaznaczmy, że przykładów uczących mamy $60000$, a przykładów testowych $-$  $10000$. Przykładowo, arkusz `sample_data/minist_train_small_csv`, który znajduje się na dysku Colaba, zawiera $20000$ obrazów i co ciekawe nie posiada wiersza nagłówkowego.

Konwersję archiwum CSV do folderu PNG jako zadanie, które nie musi się znajdować w potoku przetwarzania danych typowego projektu DNN, opisujemy i realizujemy w osobnym notatniku o nazwie [`csv2folder.ipynb`](
https://colab.research.google.com/drive/1SlYjLUQ_b4-hjkv8a7m1M-Fbkj1eCIqk?usp=sharing)
<br>Powyższy notatnik nie definuje narzędzi do konwersji arkuszy CSV reprezentujących obrazy do folderów zawierających ich postacie *oszczędne*, ale wykonuje te konwersje dla czterech konkretnych arkuszy CSV: `mnist_train_small.csv.zip`, `mnist_test_small.csv.zip`, `mnist_train.csv.zip`, oraz `mnist_test.csv.zip`.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/83/Flag_of_the_United_Kingdom_%283-5%29.svg/240px-Flag_of_the_United_Kingdom_%283-5%29.svg.png" width="20px"/><br>
In a typical DNN project, the image archive of train and test examples does not require neither conversion of image format, e.g. `JPG` vs. `PNG` nor archive type itself, e.g. `ZIP` vs. `RAR`.

In the case of `Kaggle/MNIST` archive we deal with two spreadsheets, ie. text files in `CSV` (comma separated value) format. The first spreadsheet includes train examples while the second, the test examples. The CSV spreadsheets beside the header row, includes in each consecutive row, exactly one image:
- in the first column the image label is placed, i.e. the numerical class identifier $-$ here it is the symbol of digit the image represents
- in the consecutive columns of the same row ther are values of $28\cdot 28=784$ for all pixels of the image written row-wise.

In order to make image reading faster and archive more compact on the disc, we prefer `ZIP` archive than `CSV` one. Moreover we want to keep each each in one of three image formats: `JPG`, `PNG`, and `RAW`, the latter including exactly $784$ bytes.

Since in the web there is more than one MNIST archives let us say that we are using two archives:
1. *bigger one* with $60000$ traing examples and $20000$ testing examples, i.e. the Kaggle/MNIST dataset. 
1. *smaller one* with $20000$ traing examples and $20000$ testing examples, i.e. the archive which is available in the Colab disc as default two CSVfiles: `sample_data/minist_train_small.csv` and `sample_data/minist_test.csv`.

What is interesting, the Colab version of CSV archive does not include the header row at all.

Since such a file conversion is not mandatory in the processing pipeline of DNN project,  there is separate notebook for doing this job: [csv2folder.ipynb](
https://colab.research.google.com/drive/1SlYjLUQ_b4-hjkv8a7m1M-Fbkj1eCIqk?usp=sharing)

The above Jupyter notebook does not define tools for for conversion of CSV spreadsheet representing a set of images to  more "economical" forms, however it does such conversions of the specific CSV (like those you got for free from Colaba) and creates the four data archives: `mnist_train_small.zip`, `mnist_test_small.zip`, `mnist_train.zip`, and `mnist_test.zip`.

W ciągu jednej sesji Colaba, w poszukiwaniu najlepszego modelu dla naszego problemu potrafimy wiele razy zmieniać dane uczące, detale architektury, parametry wejścia, postać funkcji strat, parametry optymalizatora.

Z jednej strony musimy więc panować nad wyborem meta-parametrów, a z drugiej nad otrzymanymi wynikami rezultatami naszych eksperymentów, w tym modelami, ich analizą i porównawczą oceną jakości. 

Eksperymenty potrafią układać w naturalne serie, ale często przechodzimy do nowej koncepcji np. architektury, która może skutkować nową serią eksperymentów. 

Pracując w Colabie musimy też liczyć się z tym, że sesja będzie zamknięta, a nasz eksperyment, a więc i seria eksperymentów będzie przerwana i nie dostaniemy czasu na zachowanie bieżącyh wyników. 

Dlatego system, który pomaga kontrolować powyższe aspekty eksperymentów obliczeniowych jest w codziennej praktyce niezbędny.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/83/Flag_of_the_United_Kingdom_%283-5%29.svg/240px-Flag_of_the_United_Kingdom_%283-5%29.svg.png" width="20px"/><br>
In one Colab session, looking for the best model for our application, we can change many times training dataset, architecture details, input parameters, loss function form, optimizer parameters, etc.

On one hand we have to control the selection of meta-parameters and on another hand the results of our experiments, their analysis, quality comparing are to be together be managed.

Experiments can be related with others creating a natural series and when we change our idea a new series begins. 

Working with Colab we are aware that our session can be closed and then our experiment interrupted without a chance to save the content neither from RAM or from Colab disc.

Therefore a system which will support controlling all above aspects is in daily practice necessary.  


####*Funkcjonalności systemu zarządzania danymi o eksperymentach obliczeniowych*<br>*Functionalities of management system for data on computing experiments*


<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/e/e9/Flag_of_Poland_%28normative%29.svg/240px-Flag_of_Poland_%28normative%29.svg.png"  width="20px"/><br>
Z powyższych scenariuszy wynika, że system `XP` wspierający zarządzanie danymi o eksperymentach może oferować następujące możliwości:
1. Inicjowanie nowej serii  eksperymentów `xp`:<br> `xp = XP()`.
1. Tworzenie obiektu informacyjnego  `e` nowego eksperymentu w serii `xp`: `e = xp.new()` lub 
1. Inicjowanie nowego eksperymentu `e` atrybutami innego eksperymentu `f`:<br> `e = xp.like(f)`.
1. Dopisywanie do eksperymentu `e` dowolnych atrybutów, np.:<br>
`e.batch_size = 512`, `e.train_accuracy = 0.98`, etc.
1. Archwizację serii eksperymentów `xp` na dysku Colaba oraz na `gdrive`:<br> `xp.save(path)` oraz `save_experiment_series(xp,exp_gfolder,?)`
1. Aktywację serii archiwizowanej `xp` z dysku Colaba oraz z `gdrive`:<br> `xp = XP.load(path)` oraz  `load_experiment_series(xp,exp_gfolder,?)`
1. Tworzenie klucza `k` dla eksperymentu w serii `xp`, uwzględniający zarówno czas rozpoczęcia eksperymentu `e` jak i jego numer seryjny `i`, np.:<br> `i = e._index_` jako numer seryjny `i`, klucz `k = e._key_` w postaci `numer seryjny: czas eksperymentu`, gdy eksperyment `e` równa się `xp.all[k]`
1. Przejście od klucza eksperymentu `k` do obiektu eksperymentu `e`:<br> `e = xp.all[k]`
1. Przejście od numeru seryjnego `i` do klucza eksperymentu `k` i dalej do obiektu eksperymentu `e`:<br> `k = xp.key(i)`, `e = xp.all[k]`
1. Przejście od obiektu eksperymentu `e` do jego klucza i numeru seryjnego `i`:<br> `k = e._key_`, `i = e._index_`
1. Przejrzyste drukowanie atrybutów eksperymentu:<br> `print(e)`
1. Konstrukcji nowego eksperymentu `e` w serii eksperymentów `xp` na podstawie indeksu  `i1` innego eksperymentu lub wielu indeksów:<br>`e = xp.like(i1)` or `e = xp.like(i1,i2,...)`
1. Przejrzyste drukowanie serii eksperymentów:<br> `xp.print()`
1. Usuwanie z serii eksperymentów `xp`, niechcianego obiektu eksperymentu `e` na podstawie jego klucza `k` lub numeru seryjnego `i` lub jego referencji `e`:<br> `xp.del_by_key(k)`, `xp.del_by_index(i)`, `xp.del_exp(e)`
1. Automatyczne wyznaczanie numeru nowej serii eksperymentów na podstawie analizy nazw plików archiwizowanych na dysku `gdrive`:<br>`new_series_id(exp_gfolder)`

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/83/Flag_of_the_United_Kingdom_%283-5%29.svg/240px-Flag_of_the_United_Kingdom_%283-5%29.svg.png" width="20px"/><br>
From the above scenarios we can conclude that a system `XP` supporting the management of data on experiments may offer the following capabilities:
1. Initiating of of new series of experiments `xp`:<br> `xp = XP()`.
1. Creating of information object  `e` for the new experiment in the series `xp`: `e = xp.new()` or
1. Initiating the new experiment `e` using attributes of another experiment `f`:<br> `e = xp.like(f)`.
1. Extending the experiment `e` by any new attribute, eg.:<br>
`e.batch_size = 512`, `e.train_accuracy = 0.98`, etc.
1. Archiving a series of experiments `xp` on Colab and `gdrive` discs:<br> `xp.save(path)` and `save_experiment_series(xp,exp_gfolder,?)`
1. Activation of archived series `xp`:<br> `xp = XP.load(path)` and `load_experiment_series(xp,exp_gfolder,?)`
1. Creating a key `k`  for the experiment series `xp`, which represents both, the starting time of the experiment `e` and its serial number `i` (aka index)   `i`, eg.:<br> `i = e._index_` as the serial number `i`, key `k = e._key_` in the form  `serial number: experiment time`, if  `e` equals  `xp.all[k]`
1. Going from the experiment key `k` to the experiment  object `e`:<br> `e = xp.all[k]`
1. Going from the serial number `i` to its key `k`, and next to the experiment object `e`:<br> `k = xp.key(i)`, `e = xp.all[k]`
1. Going from the experiment object `e` to its key `k` and its serial number `i`:<br> `k = e._key_`, `i = e._index_`
1. Constructing of a new experiment `e` in the series `xp` on the basis of another experiment index `i1` or more experiment indexes:<br>`e = xp.like(i1)` or `e = xp.like(i1,i2,...)`
1. Clear printing of experiment atributes:<br> `print(e)`
1. Clear printing of the series of experiments:<br> `xp.print()`
1. Deleting from the series `xp`, an unwanted experiment `e` by its key `k`, or its serial number `i`, or its reference `e`:<br> `xp.del_by_key(k)`, `xp.del_by_index(i)`, `xp.del_exp(e)`
1. The automatic assignment of the serial number of experiment series using the names of experiment files archived on `gdrive`:<br>`new_series_id(exp_gfolder)`

##Urozmaicenie obrazów uczących<br>Learning image augmentation

###Rodzina transformacji urozmaicających<br>Family of augmenting transformations

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/e/e9/Flag_of_Poland_%28normative%29.svg/240px-Flag_of_Poland_%28normative%29.svg.png"  width="20px"/><br>
Proces urozmaicania danych (ang. *data augmentation*) polega na losowym zaburzaniu cech obiektu wejściowego zarówno w dziedzinie przestrzennej jak i w dziedzinie wartości cech. Te losowe zmiany powinny mieć charakter lokalny, a więc są to np. niewielkie:
- obroty,
- powiększanie lub pomniejszanie globalne,
- powiększanie lub pomniejszanie w kierunki jednej z osi, np. osi $x$,
- przesunięcia w poziomie lub w pionie,
- skalowanie wartości cechy z obcięciem, gdy przekraczamy jej zakres,
- addytywne losowe zmiany wartości (szum Gaussowski czy szum ziarnisty).

Niejako osobną kategorią są:
- operacje zerowania, inaczej wygaszania lub zasłaniania cech w małych prostokątach,
- operacj symetrii osiowej, np. horyzontalnej czy wertykalnej.

Ważne jest to, że operacje urozmaicania:
- wykonują się *w locie* dla każdego przykładu uczącego i w każdej epoce, a więc zbiór uczący jest w sposób ciągły modyfikowany,
- każdy przykład podlega jednej lub ustalonej sekwencji operacji urozmaicania.

Główną motywacją takich modyfikacji jest nie tyle zwiększanie ilości przykładów uczących, a raczej wskazówka dla algorytmu uczącego, że nie należy *skupiać się* na nieistotnych szczegółach, które nie powinny wpływać na poprawność działania modelu. Jest więc to dążenie do *generalizacji wyuczonego modelu*.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/83/Flag_of_the_United_Kingdom_%283-5%29.svg/240px-Flag_of_the_United_Kingdom_%283-5%29.svg.png" width="20px"/><br>
Image data augmentation process consits of series of distortion operations both in spatial and color attribute domains. Those random changes get local and slight impact on images. For instance we apply slight:
- image rotations,
- zoom and reduce for whole image,
- zoom and reduce for along one axis, eg. $x$,
- horizontal or vertical shift,
- scaling of pixel value with clamping to ist range,
- additive random change (Gaussin noise or salt and pepper noise).

Another category of augmentation operation are:
- zeroing operation, ie. blanking or occlusion of small image rectangles,
- axis symmetry either horizontal or vertical.
Important information:
- augmentation operation are performed on fly for each learning image and for each epoch,
- each example is subject of the same sequence of operations $-$ however with different randomly selected parameters.

The main goal of such modifications is not increasing of learning data items, nut rather making a hint for training algorithm that *it should not focus* on unimportant details which should not affect the action of model. Hence it is an aim *to generalization of the trained model.*




###Moduł `imgtrans` zgodny z `tf.keras`<br>Module `imgtrans` conformed with `tf.keras`

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/e/e9/Flag_of_Poland_%28normative%29.svg/240px-Flag_of_Poland_%28normative%29.svg.png"  width="20px"/><br>

Despite that on popular DNN platforma there are well elaborated tools for image augmentation, for teaching aims I define  in the separate notebook `imgtrans.ipynb`, the Python module `imgtrans.py`, and ther the class `ImageTransforms` with the minimalistic API deriving good practices  both from Keras library (regarding the design of constructor) and PyTorch (regarding the functional notation.


<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/e/e9/Flag_of_Poland_%28normative%29.svg/240px-Flag_of_Poland_%28normative%29.svg.png"  width="20px"/><br>
Skoro nasz szablon ma działać w PyTorchu, to schemat definiowania i użycia obiektów klasy `ImageTransforms` ma takie wymogi na poziomie definiowania klas:
- w klasie typu `Dataset` dodajemy w jej konstruktorze parametr `transforms=None` i polecenie `self.transforms = transforms`, a w wymaganej metodzie `__getitem__` przed utworzeniem tensora wymaganego przez PyTorcha umieszczamy polecenie użycia zdefiniowanych transformacji w następującym kontekście:
  ```
  # ...
  img = ??? # obraz utworzony w NumPy
  if (self.transforms is not None) and (not self.validation): 
    img = self.transforms(img) # notacja funkcyjna na obiekcie
                               # transformacje wykonane też w NumPy
  img = img.astype(np.float).reshape(1,*img.shape)/255 
    # dostosowanie do zalecanego zakresu wartości pikseli: [0,1] 
    # i wymaganego dodatkowego wymiaru wsadowego: (1,rozmiary obrazu)
  return torch.tensor(img,dtype=torch.float) # i inne np. etykieta obrazu
```

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/83/Flag_of_the_United_Kingdom_%283-5%29.svg/240px-Flag_of_the_United_Kingdom_%283-5%29.svg.png" width="20px"/><br>
Since our template project is to work in PyTorch, the scheme of `ImageTransforms` objects defining and using has the following requirements:
- in the constructor of the extension class for `Dataset` we add the parameter `transforms=None` and assignment statement `self.transforms = transforms`, 
- in the interfacing method `__getitem__` just before creating the tensor expected by PyTorch we put  the command of applying the transforms defined, namely in such a context:
  ```
  # ...
  img = ??? # image created in NumPy
  if (self.transforms is not None) and (not self.validation): 
    img = self.transforms(img) # function notation for object which means applying
                               # in NumPy the operations defined
  img = img.astype(np.float).reshape(1,*img.shape)/255 
    # adjusting to recommended range of luminance: [0,1] 
    # and required batch axis: (1,image shape)
  return torch.tensor(img,dtype=torch.float) # and other items like image label
```

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/e/e9/Flag_of_Poland_%28normative%29.svg/240px-Flag_of_Poland_%28normative%29.svg.png"  width="20px"/><br>
  Uwaga: `not self.validation` dodajemy tylko wtedy, gdy chcemy korzystać z tego samego obiektu typu `Dataset` w generowaniu danych w fazie walidacji:
  - jest to flaga, którą musimy dodać w konstruktorze obok `transforms=None,validation=False)` i oczywiście
  - polecenie `self.validation = validation` obok `self.transforms = transforms`. 
  - konstruktor klasy `ImageTransforms` zawiera w swojej minmalnej wersji parametry określające zakresy losowo dokonywanych modyfikacji obrazu:

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/83/Flag_of_the_United_Kingdom_%283-5%29.svg/240px-Flag_of_the_United_Kingdom_%283-5%29.svg.png" width="20px"/><br>
Note: `not self.validation` is included only if we want to use the same class extending `Dataset` for validation phase:
- it is a flag which is to be next to `transforms=None, validation=False)
- the flag  is saved `self.validation = validation` just next to `self.transforms = transforms`. 
- the constructor of `ImageTransforms` class includes in its minimalistic form, the following parameters which define ranges for selecting parameter controllin image modifications:
  ```
def __init__(self,rotation_range = None, zoom_range = None,
```

````
%%cython
cimport cython
cimport numpy as np
def interpolate_bilinearly_cy(double ax, double bx, double cx,\
  double ay, double by, double cy, Py_ssize_t yres, Py_ssize_t xres, Py_ssize_t cres,\
        unsigned char[::1] buf_in, unsigned char[::1] buf_out,): 
  cdef double xres1 = xres-1.0, yres1 = yres-1.0
  cdef double x, y, value
  cdef Py_ssize_t sx = cres, sy= cres*xres, sc= 1
  cdef Py_ssize_t i, j, k, iA, iB, jA, jC
  cdef double f_xa, f_xb, f_ya, f_yc
  cdef Py_ssize_t ofs, ksc
  k = 0; ofs = 0
  while k<cres:
    j = 0; ksc = ofs = k*sc; x = cx; y = cy
    while j<yres:
      i = 0; ofs = j*sy+ksc; x = j*bx+cx; y = j*by+cy
      while i<xres:
        #x = ax*i+bx*j+cx; y = ay*i+by*j+cy # liczymy przyrostowo
        xx = 0.0 if x<0.0 else xres1 if x>xres1 else x
        iA = int(xx); iB = iA+1 if (xx< xres1 and xx>0) else iA; f_xa = xx-iA; f_xb = iB-xx
        yy = 0.0 if y<0.0 else yres1 if y>yres1 else y
        jA = int(yy); jC = jA+1 if (yy<yres1 and yy>0) else jA; f_ya = yy-jA; f_yc = jC-yy
        value = f_xb*f_yc*buf_in[iA*sx+jA*sy+ksc] +\
                f_xa*f_yc*buf_in[iB*sx+jA*sy+ksc] +\
                f_xb*f_ya*buf_in[iA*sx+jC*sy+ksc] +\
                f_xa*f_ya*buf_in[iB*sx+jC*sy+ksc]  
        value = 0.0 if value<0.0 else 255. if value>255. else value+0.5
        buf_out[ofs] = int(value)
        i += 1; ofs += sx; x += ax; y += ay
      j += 1; ofs += sy; x += bx;  y += by
    k += 1; ofs += sc
````

````
import torch
from torch.utils.data import Dataset
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
````

````
class DataProvider(Dataset):
  def __init__(self,path,ext='raw', # 'jpg', 'png'
               mode='train', # 'test', 'train', 'val'
               val=10, # procent danych treningowych branych do walidacji
               transforms=None, # urozmaicanie danych
               shape=None, #wymagane rozmiary gdy ext='raw'
               force_color=False,
               ):
    self.path = path; self.ext = ext; self.mode = mode; self.force_color = force_color
    self.transforms = transforms; self.shape = shape; #print('shape:',shape)
    with open(path+'/labels.pck','rb') as f: self.labels = pickle.load(f)
    n = len(self.labels);  self.base = n-int(n*val/100.)
    self.idxes = list(range(n)); random.shuffle(self.idxes)
    self.nlabels = n if mode=='test' else n-self.base if mode=='val' else\
                    self.base # if mode=='train'
  def __len__(self): return self.nlabels
  def __getitem__(self,idx): return __getitem__(self,idx)
````

Kolejną rzeczą wartą komentarza jest sposób liczenia liczby przykładów uczących i liczby przykładów walidacyjnych. Wynika on z takiej konwemcji:
- indeksy $0,\dots,b-1$ to indeksy przykładów uczących, a

Another comment refers on relationship between the number of learning and validating examples. It follows from the following assumption:
- $0,\dots,b-1$ $-$ indexes of learning examples

In the code we get:

$b\mapsto$ `self.base`,<br>
$n\mapsto$ `len(self.labels)`.


````
def __getitem__(self, idx):
  idx_true = self.idxes[self.base+idx] if self.mode=='val' else self.idxes[idx]
  fpath = self.path+'/'+str(idx_true)+'.'+self.ext
  if self.ext!='raw': img = imread(fpath); #print('img shape:',img.shape)
  else:
    with open(fpath,'rb') as f: 
      img = np.frombuffer(f.read(),dtype=np.uint8).reshape(self.shape)
  if (self.transforms is not None) and self.mode=='train': img = self.transforms(img) 
  if self.force_color and len(img.shape)==2: img = np.stack([img,img,img],axis=-1)
  if len(img.shape)==3: img = img.transpose([2,0,1])
  shp = img.shape; toshape = (1,*shp) if len(shp)==2 else shp
  img = img.astype(np.float32).reshape(*toshape)/255 
  image = torch.from_numpy(img).to(device)
  label = torch.tensor(int(self.labels[idx_true]),dtype=torch.long)
  return label,image
````

###Losowy wsadowy przegląd danych uczących<br>Batch random scan of learning data

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/e/e9/Flag_of_Poland_%28normative%29.svg/240px-Flag_of_Poland_%28normative%29.svg.png"  width="20px"/><br>
Formowanie grup danych w postaci tzw. mini-batchów, jest wymagane ze względu na efektywnośc procesu uczenia na kartach graficznych. Ponieważ wymagana jest również losowość procesu tworzenia wsadów, to wystarczy by losowo generowane indeksy elementów danych były przekazywane do ich dostawcy w postaci obiektu z klasy `DataProvider`.

W PyTorchu losowy wsadowy przegląd danych rozwiązuje obiekt z klasy `DataLoader`, który oferuje interfejs iteratora który dostarcza na życzenie kolejne losowe grupy danych, a tworzy je wywołując metodę `__getitem__` z klasy typu `Dataset`, a więc np. z naszej klasy `DataProvider`. By to było możliwe w konstruktorz `DataLoader` rejestruje właśnie naszego dostawcę pojedyńczych elementów danych.

Poniżej mamy przykład tworzenia tego ważnego ogniwa między zbiorami danych uczących i testujących, a pętlą uczenia, w której przegląd tych danych sie dokonuje.

Grouping of data items into so called mini-batches is necessary for effective model learning on graphics cards. Since in each scan of all traing data items, i.e. during an epoch, mini-batch should be randomly selected (the requirement of stochastic optimization) it is enough to request the mini-batch by sending the data random indexes. 

In PyTorch the random batch scan of the whole traing dataset is solved by the object from `DataLoader` class which offers an iterator to select random indexes and call for them `__getitem__` method of user provided class.

The code below shows how by registering `DataProvider` objects  having this interface, the gap between datasets and training loop is filled.

Komponent splotowy może być opcjonalnie poprzedzony jednostką regularyzacyjną  typu `dropout`, ale kolejne jednostki to sekwencja jednostek splotowych. 

Parametry komponentu splotowego przekazywane są na liście rozmiarów maski splotu oraz na opcjonalnych listach kroków próbkowania (*stride*) i flag uzupełnienia *padding*. Zakładamy, że liczba filtrów, a więc map cech liczonych w każdej jednostce splotowej danego komponentu jest taka sama. 

Po serii komponentów splotowych mamy opcjonalny *dropout* jedną jednostkę splotową o filtrach, których maski całkowicie pokrywają dziedzinę pikselową tensora otrzymanego na wyjściu ostatniego komponentu splotowego. 

Ostatnia warstwa naszej sieci to wspomniana jednostka dopasowująca wyjściowy wektor cech głębokich do liczby klas.

Opisaną  strukturę sieci typu `PureConv2dNet` możemy prześledzić na sieci $\cl{N}_3$ z dwoma komponentami typu `Conv2dComponet`, a w każdym z nich mamy po trzy sploty.

The next networks are created according to the common scheme which aims to use purely convolutional components of type `Conv2dComponent` eith only exception for the ending affine layer which computes the scores for classes. 

Each convolutional component can be optionally preceded by the feature regularization unit of type `Dropout` while next layers create the sequence of convolutional units.

Parameters for convolutional component are passed as one mandatory list of *kernel* sizes*and two optional lists, one for subsampling *strides*  and the second for flags of *padding*. We assume that the number of filters and therefore the number of feature maps for each convolutional unit of the component is the same. 

After a series of convolutional components we have an optional  *dropout* , the single convolutional unit with the mask covering the spatial domain of its input tensor.

The last layer is the mentioned affine unit which matches the feature vector size to the number of classes.

The implementation of the universal *purely convolutional* network, as the matter of fact, is not more complex than rather special network $\cl{N}_0$. It needs only handling the lists of parameters.

````
class Conv2dComponent(nn.Module):
  def __init__(self,nf_in,nf_out,klist,slist=None,plist=None):
    super(Conv2dComponent,self).__init__()
    self.convs = nn.ModuleList(); self.bns = nn.ModuleList()
    for i,k in enumerate(klist):
      nf = nf_out if i else nf_in
      s = slist[i] if slist else 1 
      p = k-1 if (plist and plist[i]) else 0 
      self.convs.append(nn.Conv2d(nf,nf_out,k,stride=s,padding=p//2))
      self.bns.append(nn.BatchNorm2d(nf_out))
  def forward(self,x):
    for i,conv in enumerate(self.convs): 
      x = self.bns[i](F.relu(conv(x)))
    return x
````

````
class PureConv2dNet(nn.Module):
  def __init__(self,flist,klists,outres, # = 4
               slists=None,plists=None,dlist=None,nclasses=10):
    super(PureConv2dNet,self).__init__()
    self.convcomps = nn.ModuleList(); self.d1s = nn.ModuleList()
    for i,klist in enumerate(klists):
      slist = slists[i] if slists else None
      plist = plists[i] if plists else None
      self.convcomps.append(Conv2dComponent(flist[i],flist[i+1],klist,
                                            slist=slist,plist=plist))
      if dlist and dlist[i]>0: self.d1s.append(nn.Dropout(dlist[i]))
      else: self.d1s.append(None)
    self.d2 = nn.Dropout(dlist[-1]) if (dlist and dlist[-1]>0) else None
    self.conv = nn.Conv2d(flist[-2],flist[-1],outres)
    self.fc = nn.Linear(flist[-1], nclasses)
  def forward(self,x):
    for i,convcomp in enumerate(self.convcomps):
      x = self.d1s[i](x) if self.d1s[i] else x; x = convcomp(x) #; print('shape:',x.shape)
    x = self.d2(x) if self.d2 else x; x = self.conv(x) #; print('shape:',x.shape)
    x = torch.flatten(x, start_dim=1) #; print('shape:',x.shape)
    y = self.fc(x); return y
````

####Funkcja `train`<br>Function `train`

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/e/e9/Flag_of_Poland_%28normative%29.svg/240px-Flag_of_Poland_%28normative%29.svg.png"  width="20px"/><br>
Funkcja `train` w trakcie budowania modelu informuje o poziomie błędu modelowania co wybraną liczbę wsadów trenujących oraz po każdej epoce o średnim błędzie modelowania na zbiorze walidacyjnym.  

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/83/Flag_of_the_United_Kingdom_%283-5%29.svg/240px-Flag_of_the_United_Kingdom_%283-5%29.svg.png" width="20px"/><br>
Function `train` while building the model it informs on the modelling error level, after a selected number of mini-batches (learning ones) and after each epoch it informs about the mean error  of modelling computed on the validation set.  

````
def train(net,tgen,vgen,loss_function,optimizer,scheduler,batch_size,num_epochs):
  h = {'tloss_e': [], 'vloss_e': [], 'tacc_e': [], 'vacc_e': [], }; max_vacc = -1. 
  for epoch in range(num_epochs):
    nitems = 0; tloss = 0.0; tacc = 0
    for i,(labels,images) in enumerate(tgen):
      labels = labels.to(device)
      outputs = net(images); optimizer.zero_grad()
      loss = loss_function(outputs, labels); loss.backward()
      optimizer.step()
      _, predicted = torch.max(outputs,1); tacc += (predicted == labels).sum()
      nitems += labels.size(0); tloss += loss
      if (i+1) % 500 == 0:
        print('epoch [{}/{}], batch [{}/{}], current mean loss: {:.5f}'\
        .format(epoch+1, num_epochs, i+1,len(tgen),tloss/(i+1)))
    scheduler.step()
    vloss_n, vacc_n = validate(net,loss_function,vgen)
    h['vloss_e'].append(vloss_n); h['vacc_e'].append(vacc_n)
    tloss_n, tacc_n = validate(net, loss_function,tgen)
    #tloss_n = tloss/len(tgen); tacc_n = tacc/nitems
    h['tloss_e'].append(tloss_n); h['tacc_e'].append(tacc_n)
    if vacc_n>max_vacc and epoch>0: 
      save_model(ex,net,optimizer,ioc_mod_folder,h,ioc); max_vacc = vacc_n
    print(('epoch [{}/{}], mean loss->(train: {:.5f}, val: {:.5f}), '+\
          'success rate->(train: {:.2f}%, val: {:.2f}%)').format(epoch+1, num_epochs,\
          tloss_n,vloss_n,100*tacc_n,100*vacc_n))
  return h    
````

####Funkcja `validate`<br>Function `validate`

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/e/e9/Flag_of_Poland_%28normative%29.svg/240px-Flag_of_Poland_%28normative%29.svg.png"  width="20px"/><br>
Funkcja `validate` liczy średnią wartość funkcji strat (`loss_function`) na zbiorze przykładów walidacyjnych przeglądanych iteratorem `gen`.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/83/Flag_of_the_United_Kingdom_%283-5%29.svg/240px-Flag_of_the_United_Kingdom_%283-5%29.svg.png" width="20px"/><br>
The function `validate` computes the mean value of loss function for the validating examples which are scanned using the iterator `gen`.

````
def validate(net,loss_function,gen):
  with torch.no_grad():
    vloss = 0.0; correct = 0; nitems = 0
    for labels,images in gen:
      labels = labels.to(device); nitems += labels.size(0)
      outputs = net(images)
      loss = loss_function(outputs, labels); vloss += loss.data.item()
      _, predicted = torch.max(outputs,1); correct += (predicted == labels).sum()
    return vloss/len(gen), correct.data.item()/nitems
````

####Funkcja `test`<br>Function `test`

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/e/e9/Flag_of_Poland_%28normative%29.svg/240px-Flag_of_Poland_%28normative%29.svg.png"  width="20px"/><br>
Funkcja `test` wyznacza miarę dokładności modelowania - w tym przypadku średnią stopu sukcesu klasyfikatora cyfr.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/83/Flag_of_the_United_Kingdom_%283-5%29.svg/240px-Flag_of_the_United_Kingdom_%283-5%29.svg.png" width="20px"/><br>
The function `test` determines the metrics of modelling accuracy $-$ in this case the success rate.

````
def test(net,gen):
  correct = 0; nitems = 0
  with torch.no_grad():
    for labels,images in gen:
      labels = labels.to(device); nitems += labels.size(0)
      outputs = net(images)
      _, predicted = torch.max(outputs,1); correct += (predicted == labels).sum()
  return correct/nitems
````

Gdy chcemy kontynuować budowę modelu w tej samej lub nowej sesji, to przydatne będą  funkcje `save_model` oraz `load_model`, które korzystają z mechanizmów obsługi tej funkcjonalności w PyTorchu, podczas gdy tworzą stosowne ścieżki dostępu do modeli i opisów eksperymentów, w ramach których te modele powstały. 
1. Gdy więc nasz bieżący eksperyment jest utworzony w ramach istniejącej serii, to jej aktualna postać jest serializowana najpierw do pliku `experiments.pck` na dysku Colaba, a potem transferowana na gdrive. 
1. Gdy jest to nowa seria, to otrzymuje ona kolejny numer $n=?$ na podstawie przeglądu istniejących numerów i pod nazwą `series-n.pck` trafia na dysk gdrive.
1. Gdy chcemy zapisać model utworzony w eksperymencie $j$ w serii $i$, to wpiszemy go obok pliku serii eksperymentów do folderu `models-i/model-j.pck` za pośrednictwem jego serializowanej formy w opliku `model.pck` na dysku Colaba. 

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/8/83/Flag_of_the_United_Kingdom_%283-5%29.svg/240px-Flag_of_the_United_Kingdom_%283-5%29.svg.png" width="20px"/><br>
1. If we want to continue the model training in the same or any new session, then the utility functions `save_model` and  `load_model` in PyTorch will be used to keep information on the current state of model, its optimizer, and addtionally on parameters of the experiment which has led to those states. 

1. If our eperiment was created within a series of experiments then the whole series is serialized (pickled) and saved to the Colab file `experiments.pck`, and next transferred to gdrive. 

1. If the session is new one then it receives the serial number $n$, based on reiew of existing numbers and with the name `series-n.pck` it is uploaded to `gdrive`.

1. If we want to save the model in the experiment  $j$ in the series $i$ then we write this as the the file  `models-i/model-j.pck`. The folder `models-i` is created manually by the user just next to file `series-i.pck`.

````
def save_model(ex,net,optimizer,ioc_mod_folder,h,ioc):
  torch.save({ 
      'experiment': ex,
      'history': h,
      'net state dict': net.state_dict(),
      'optimizer state dict': optimizer.state_dict()
    }, 'model.pck')
  try:
    model_path = ioc_mod_folder+'models-'+str(ex.series_id)+'/model-'+str(ex._index_)+'.pck'
    ioc.cg('model.pck',model_path)
  except: raise Exception('Path '+model_path+\
                          ' is wrong\nit seems that model folder is not created yet!')
def load_model(net,series_id,exp_index,ioc_mod_folder,ioc):
  model_path = ioc_mod_folder+'models-'+str(series_id)+'/model-'+str(exp_index)+'.pck'
  print('ioc model folder:',ioc_mod_folder)
  ioc.gc(model_path,'model.pck')
  cpoint = torch.load('model.pck')
  net.load_state_dict(cpoint['net state dict'])
  optimizer.load_state_dict(cpoint['optimizer state dict'])
  return cpoint['experiment'], cpoint['history']
````
